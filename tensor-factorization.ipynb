```python
directory = "../data/trade_data/data/"
df_names = []
for filename in os.listdir(directory):
    df_name = filename.replace(".csv", "")
    df_names.append(df_name)

print(len(df_names))
print(df_names)
```

```python
# assign dataframes to each csv

# columsn to read in
cols = ["Classification", "Year", "Trade Flow", "Trade Value (US$)"]

# creating each dataframe for file in data directory
for i in range(0,len(df_names)):
    df_name = df_names[i]
    print(df_name)
    
    filepath = os.path.join(directory, df_names[i] + ".csv")
    
    # detect the files encoding (not all are utf-8)
    with open(filepath, 'rb') as f:
        enc = chardet.detect(f.read())
    
    #this programmatically assigns a df variable name
    locals()[df_name] = pd.read_csv(filepath,  encoding = enc['encoding'], usecols = cols)
    
```
